```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import *

def makeData(N,noise):
  xs=np.random.random(N)*2
  ys=np.sin(4*xs)+noise*np.random.randn(N)
  return xs,ys

xs,ys=makeData(300000,.1)
plt.scatter(xs,ys)
plt.show()

xsdata=xs.reshape(-1,1)#make the 30-element array xs into a 30x1 feature matrix

pipe=make_pipeline(
    PolynomialFeatures(degree=4),
    StandardScaler(),
    Ridge(alpha=0.01)
    )

print(xsdata.shape)
pipe.fit(xsdata,ys)

testXs=np.linspace(0,2).reshape(-1,1)
testYs=pipe.predict(testXs)
plt.scatter(xs,ys)
plt.plot(testXs,testYs)
plt.show()
```
